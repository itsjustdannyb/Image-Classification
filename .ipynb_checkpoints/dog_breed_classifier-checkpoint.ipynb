```python
import os
from PIL import Image # for image handling
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Processing

```python
import numpy as np
path = r"/content/drive/MyDrive/datasets - images/dog-breed-images/dataset"

def show_paths(path, size=(125,125)):
    images = [] # numerical
    labels = []
    dog_breeds = os.listdir(path)

    for dog_breed in dog_breeds:
        dog_folder = os.path.join(path, dog_breed)
        if not os.path.isdir(dog_folder):
            continue

        for dog in os.listdir(dog_folder):
            img_path = os.path.join(dog_folder, dog)
            img = Image.open(img_path).convert("RGB") # make image RGB
            # img = img.resize(size, Image.ANTIALIAS) #antialias tries to retain image quality when resized
            img = img.resize(size, Image.Resampling.LANCZOS) #Image.Resampling.LANCZOS updated version of Image.ANIALIAS

            img_array = np.array(img) # height, width and channels

            images.append(img_array)
            labels.append(dog_breed)

    images = np.array(images)
    labels = np.array(labels)

    return images, labels

X, y = show_paths(path)

print(X.shape)
print(y.shape)
```

## Normalize data

```python
X_scaled = X / 225.0
X
```

### Label Encode data

```python
len(np.unique(y))
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
y_encoded
```

## one hot encode labels

```python
from tensorflow.keras.utils import to_categorical
y_encoded = to_categorical(y_encoded)
y_encoded
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

ratio_train = 0.6
ratio_val = 0.2
ratio_test = 0.1

X_tmp,X_test,y_tmp,y_test = train_test_split(X_scaled,y_encoded,test_size=ratio_test, random_state=42)
X_train,X_val,y_train,y_val = train_test_split(X_tmp,y_tmp,test_size=0.2/(1-ratio_test), random_state=42)

print(X_train.shape)
print(y_train.shape)
print(X_val.shape)
print(y_val.shape)
print(X_test.shape)
print(y_test.shape)
```

##Model Training

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
```

```python
model = Sequential([
    Flatten(input_shape=(125, 125, 3)),
    Dense(units=32, activation = 'relu'),
    Dense(units=16, activation='relu'),
    Dense(units=10, activation='softmax')
], name = "nn-approach")
```

```python
optimizer = tf.keras.optimizers.Adam()
loss = tf.keras.losses.CategoricalCrossentropy()

model.compile(
    optimizer = optimizer,
    loss = loss,
    metrics = ['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs = 50, validation_data = (X_val, y_val))
```

```python
model.summary()
```

## Convolutional Neural Network Approach

```python

cnn_model = Sequential([
    Conv2D(filters = 32, kernel_size = (3, 3), activation = 'relu', input_shape = (125, 125, 3)),
    Conv2D(filters = 64, kernel_size = (3, 3), activation = 'relu'),
    Conv2D(filters = 128, kernel_size = (3, 3), activation = 'relu'),
    Flatten(),
    Dense(units = 10, activation = 'softmax')

], name="cnn_approach")
```

```python
optimizer = tf.keras.optimizers.Adam()
loss = tf.keras.losses.CategoricalCrossentropy()

cnn_model.compile(
    optimizer = optimizer,
    loss = loss,
    metrics = ['accuracy']
)
```

```python
cnn_model.fit(X_train, y_train, epochs = 30, validation_data = (X_val, y_val))
```

```python
cnn_model.summary()
```

```python
history = model.history.history
history = pd.DataFrame(history)
history
```

```python
plt.figure(figsize=(12, 9))
sns.lineplot(history['loss'], label="train_loss")
sns.lineplot(history['val_loss'], label="val_loss")
plt.legend()
plt.title("Cost Function for a Multi-layer Perceptron Approach")
```

```python
cnn_history = cnn_model.history.history
cnn_history = pd.DataFrame(cnn_history)
cnn_history
```

```python
plt.figure(figsize=(12,9))
sns.lineplot(cnn_history['loss'], label = "train_loss")
sns.lineplot(cnn_history['val_loss'], label = "validation_loss")
plt.title("Cost Function for a CNN approach")
plt.legend()
```
